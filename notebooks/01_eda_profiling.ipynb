```python
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 60)
```

```python
raw_dir = "../data/raw"
files = sorted(os.listdir(raw_dir))

print(f"Total files found: {len(files)}\n")
for f in files:
    print(f)
```

```python
print("=" * 70)
print("COLUMN AUDIT ACROSS ALL FILES")
print("=" * 70)

column_map = {}

for filename in files:
    path = os.path.join(raw_dir, filename)
    try:
        df = pd.read_csv(path, nrows=0, encoding='latin1')  # just headers
        cols = list(df.columns)
        column_map[filename] = cols
        print(f"\n📄 {filename}")
        print(f"   Columns ({len(cols)}): {cols}")
    except Exception as e:
        print(f"\n❌ {filename} — ERROR: {e}")
```

```python
print("=" * 70)
print("COLUMNS THAT DIFFER ACROSS FILES")
print("=" * 70)

all_columns = set()
for cols in column_map.values():
    all_columns.update([c.strip().lower() for c in cols])

print(f"\nAll unique column names found across all files:\n")
for col in sorted(all_columns):
    print(f"  - {col}")
```

```python
print("=" * 70)
print("FILE SIZES")
print("=" * 70)

summary = []

for filename in files:
    path = os.path.join(raw_dir, filename)
    try:
        df = pd.read_csv(path, encoding='latin1')
        summary.append({
            'file': filename,
            'rows': len(df),
            'columns': len(df.columns)
        })
    except Exception as e:
        summary.append({'file': filename, 'rows': 'ERROR', 'columns': str(e)})

summary_df = pd.DataFrame(summary)
print(summary_df.to_string(index=False))
print(f"\nTotal rows across all files: {summary_df['rows'].sum():,}")
```

```python
df_april = pd.read_csv("../data/raw/april_2015.csv", encoding='latin1')

print("SHAPE:", df_april.shape)
print("\nCOLUMNS:")
print(df_april.columns.tolist())
print("\nFIRST 5 ROWS:")
df_april.head()
```

```python
print("DATA TYPES:")
print(df_april.dtypes)

print("\nNULL COUNTS:")
print(df_april.isnull().sum())

print(f"\nNull % per column:")
print((df_april.isnull().mean() * 100).round(1))
```

```python
# Make sure we find the amount column regardless of exact name
amount_col = [c for c in df_april.columns if 'amount' in c.lower()][0]
print(f"Amount column: '{amount_col}'\n")

# Convert to numeric
df_april[amount_col] = pd.to_numeric(df_april[amount_col], errors='coerce')

print("Amount Statistics:")
print(df_april[amount_col].describe())

print(f"\nNegative values (refunds): {(df_april[amount_col] < 0).sum()}")
print(f"Zero values: {(df_april[amount_col] == 0).sum()}")
print(f"Positive values: {(df_april[amount_col] > 0).sum()}")
print(f"Null/unparseable: {df_april[amount_col].isnull().sum()}")

print(f"\nTotal gross spend: £{df_april[df_april[amount_col]>0][amount_col].sum():,.2f}")
print(f"Total refunds:     £{df_april[df_april[amount_col]<0][amount_col].sum():,.2f}")
print(f"Net spend:         £{df_april[amount_col].sum():,.2f}")
```

```python
supplier_col = [c for c in df_april.columns if 'supplier' in c.lower()][0]
print(f"Supplier column: '{supplier_col}'\n")

print(f"Unique suppliers: {df_april[supplier_col].nunique()}")
print(f"\nTop 10 suppliers by spend:")
print(
    df_april.groupby(supplier_col)[amount_col]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .apply(lambda x: f"£{x:,.2f}")
)
```

```python
for col in df_april.columns:
    if 'expense' in col.lower() or 'expenditure' in col.lower():
        print(f"\n📌 Column: '{col}'")
        print(f"   Unique values: {df_april[col].nunique()}")
        print(df_april[col].value_counts().head(10))
```

```python
date_col = [c for c in df_april.columns if 'date' in c.lower()][0]
print(f"Date column: '{date_col}'\n")

df_april[date_col] = pd.to_datetime(df_april[date_col], errors='coerce', dayfirst=True)
print("Date range:")
print(f"  Min: {df_april[date_col].min()}")
print(f"  Max: {df_april[date_col].max()}")
print(f"\nNull dates: {df_april[date_col].isnull().sum()}")
```

```python
print(f"Total rows: {len(df_april)}")
print(f"Fully duplicate rows: {df_april.duplicated().sum()}")

# Check transaction number duplicates if column exists
txn_cols = [c for c in df_april.columns if 'transaction' in c.lower() or 'number' in c.lower()]
if txn_cols:
    txn_col = txn_cols[0]
    print(f"\nTransaction Number column: '{txn_col}'")
    print(f"Duplicate transaction numbers: {df_april[txn_col].duplicated().sum()}")
    print(f"\nExample duplicates:")
    dupes = df_april[df_april[txn_col].duplicated(keep=False)].sort_values(txn_col)
    print(dupes[[txn_col, supplier_col, amount_col]].head(8))
```
